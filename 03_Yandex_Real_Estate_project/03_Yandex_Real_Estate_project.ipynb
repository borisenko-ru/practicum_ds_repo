```python
df['cityCenters_nearest'] = df['cityCenters_nearest'].fillna(0)
df_cc_median = df.fillna(0).pivot_table(index = ['locality_name','locality_name_id'], values = 'cityCenters_nearest', aggfunc = ['median'])
df_cc_median = pd.DataFrame(df_cc_median.to_records())
df_cc_median = df_cc_median.rename(columns = {"('median', 'cityCenters_nearest')": 'cityCenters_median'}, inplace = False)
df_dict['cityCenters_median'] = df_dict['locality_name_id'].map(df_cc_median.set_index('locality_name_id')['cityCenters_median'])
```

- *df_dict* (рассчетные значения населенных пунктов с количеством объявлений 10 и более)

Оставшиеся единичные пропуски по расстоянию до аэропорта заполним медианными значениями, а также пропущенные значения по времени размещения объявления.

Опишем следующие параметры: площадь, цена, число комнат, высота потолков и время продажи. Уберём редкие и выбивающиеся значения и посмотрим, что изменится. Создадим функцию для их удаления в целях лучшей визуализации.

```python
def del_outliers(df, col, add_boxplot = 0, x = 10, y = 10):
    btm_quartile = df[col].quantile(0.25)                 
    top_quartile = df[col].quantile(0.75)
    interquartile = top_quartile - btm_quartile
    true_data_range = (df[col] >= (btm_quartile - 1.5*interquartile)) & (df[col] <= (top_quartile + 1.5*interquartile))
    if add_boxplot == 1:
        return df.loc[true_data_range].boxplot([col], figsize=(x,y))
    else:
        return df.loc[true_data_range]
```

#### Анализ квартир в центре

Выделим количество объявлений в центре.

```python
cityCenters_area_apt = pd.DataFrame(spb_price_per_km.query('cityCenters_nearest_km <= 10'))
len(cityCenters_area_apt)
```

Изучим общие параметры полученной выборки.

```python
print('Средняя площадь: {:.0f} м\u00b2'.format(cityCenters_area_apt.query(total_area_query)['total_area'].median()), end='\n')
print('Средняя цена: {:.0f} рублей'.format(cityCenters_area_apt.query(last_price_query)['last_price'].median()), end='\n')
print('Среднее число комнат: {:.0f}'.format(cityCenters_area_apt.query(rooms_query)['rooms'].median()), end='\n')
print('Средняя высота потолков: {:.2f} м'.format(cityCenters_area_apt.query(ceiling_height_query)['ceiling_height'].median()), end='\n')
```

Суммируем срез стоимости по всем параметрам для вывода общей корреляции.

```python
price_query = f'{total_area_query} and {rooms_query} and {ceiling_height_query} and {days_exposition_query}'
price = cityCenters_area_apt.query(price_query)

features = [
    'last_price',
    'cityCenters_nearest_km',
    'total_area',
    'rooms',
    'ceiling_height',
    'publication_dayofweek',
    'publication_month',
    'publication_year'
]

price_features_corr = price[features].corr()
price_features_corr.sort_values('last_price', ascending=False)
```

```python
price.pivot_table(index='floor_type', values='last_price').sort_values(by='last_price')
```

Однако, в центре чем выше этаж, тем дороже квартира. Всем охота полюбоваться на город с высоты.

```python
cmap = plt.get_cmap('coolwarm')
f = plt.figure(figsize=(5, 5))

plt.matshow(price_features_corr, cmap=cmap, fignum=f.number)
plt.xticks(range(price_features_corr.shape[1]), price_features_corr.columns, fontsize=9, rotation=90)
plt.yticks(range(price_features_corr.shape[1]), price_features_corr.columns, fontsize=9)

cb = plt.colorbar()
cb.ax.tick_params(labelsize=9)
```

Исходя из результатов проведенной корреляции очевидна прямая зависимость стоимости квартиры от расположения относительно центра и общей площади. Также можно проследить легкую зависимость цены от высоты потолков и количества комнат, но это скорее следствие расположения квартиры и ее общей площади.
